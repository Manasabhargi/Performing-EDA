### Performing EDA on visa dataset

#### What is Exploratory Data Analysis (EDA) ?
Ans:- It is performed for a better understanding of data aspects like:<br>
      -> Features of data<br>
      -> variables and relationships that holds between them<br>
      -> identifying which variables are important<br>

#### What are the Different methods of EDA?<br>
Ans:- 1. Descriptive Statistics<br>
      2. Grouping Data<br>
      3. ANOVA test<br>
      4. Correlation

### Data Visualization

```python
fig, axs = plt.subplots(1,2, figsize = (27, 17) )
plt.subplot(121)
sns.histplot(data = df, x='yr_of_estab',bins=30,kde=True, color = 'g')
plt.subplot(122)
sns.histplot(data=df,x='yr_of_estab', kde=True, hue='region_of_employment')
plt.show()
```

#### Insight
Region of employment is more in north east zone according to the histogram shown above.<br>
Though the establishment has begun during 1800 itself, the employment tends to increase during the year 2000


#### Violin plot representation with respect to experience

```python

plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('Has Experience')
sns.violinplot(y='no_of_employees',data=df,color='red',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('Need Training')
sns.violinplot(y='prevailing_wage',data=df,color='green',linewidth=3)
```

#### BIVARIATE ANALYSIS

```python
gender_group = df.groupby('case_status').mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Year of Establishment','Count of Employees']


female_scores = [gender_group['yr_of_estab'][0], gender_group['no_of_employees'][0]]
male_scores = [gender_group['yr_of_estab'][1], gender_group['no_of_employees'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Denied')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Certified')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Year of establishment vs region of employment with case status", fontweight='bold')
plt.legend()
plt.show()
```

#### Multicariate Analysis using pieplot

```python
df
```

```python
plt.rcParams['figure.figsize'] = (30, 12)

plt.subplot(1, 5, 1)
size = df['continent'].value_counts()
labels = 'Asia', 'Africa', 'North America', 'Europe', 'South America', 'Oceania'
color = ['red','green','blue','yellow','orange','brown']


plt.pie(size, colors = color, labels = labels,autopct = '.%2f%%')
plt.title('Continent', fontsize = 20)
plt.axis('off')



plt.subplot(1, 5, 2)
size = df['education_of_employee'].value_counts()
labels = 'High School', 'Master`s','Bachelor`s','Doctorate'
color = ['red', 'green', 'blue', 'cyan']

plt.pie(size, colors = color,labels = labels,autopct = '.%2f%%')
plt.title('education_of_employee', fontsize = 20)
plt.axis('off')



plt.subplot(1, 5, 3)
size = df['region_of_employment'].value_counts()
labels = 'West', 'Northeast','South','Midwest','Island'
color = ['red','green','cyan','blue']

plt.pie(size, colors = color,labels = labels,autopct = '.%2f%%')
plt.title('Region of Emplymnet', fontsize = 20)
plt.axis('off')


plt.subplot(1, 5, 4)
size = df['case_status'].value_counts()
labels = 'Certified', 'Denied'
color = ['red','green']

plt.pie(size, colors = color,labels = labels,autopct = '.%2f%%')
plt.title('Case Status', fontsize = 20)
plt.axis('off')



plt.tight_layout()
plt.grid()

plt.show()
```

#### UNIVARIATE Analysis of the feature based on the emplyees posted in different regions(region_of_emplyment)

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['region_of_employment'],data=df,palette = 'bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x = df['region_of_employment'].value_counts(),labels=df['region_of_employment'].value_counts().index,explode=[0.1,0,0,0,0],autopct='%1.1f%%',shadow=True)
plt.show() 
```

#### UNIVARIATE ANALYSIS based on the feature of education

```python
plt.rcParams['figure.figsize'] = (15, 9)
plt.style.use('fivethirtyeight')
sns.countplot(df['education_of_employee'], palette = 'Blues')
plt.title('Comparison of Education', fontweight = 30, fontsize = 20)
plt.xlabel('Degree')
plt.ylabel('count')
plt.show()

```

#### BIVARIATE ANALYSIS to check whether the education has any impact on the wages of the employees

```python
df.groupby('education_of_employee').agg('mean').plot(kind='barh',figsize=(10,10))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 9)
plt.style.use('seaborn-talk')
sns.countplot(df['unit_of_wage'], palette = 'PuBu')
plt.title('Comparison of wages provided', fontweight = 30, fontsize = 20)
plt.xlabel('Unit of Wage')
plt.ylabel('count')
plt.show()
```
